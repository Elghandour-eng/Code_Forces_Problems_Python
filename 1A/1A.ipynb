
## 1A - Theare Square

<p>Theatre Square in the capital city of Berland <br>
has a rectangular shape with the size n × m meters.<br>
On the occasion of the city's anniversary, a decision was<br>
taken to pave the Square with square granite flagstones.<br>
Each flagstone is of the size a × a.<br>
What is the least number of flagstones needed to pave the Square? <br>
It's allowed to cover the surface larger than the Theatre Square, <br>
but the Square has to be covered. It's not allowed to break the flagstones. <br>
The sides of flagstones should be parallel to the sides of the Square.</p>

Link : https://codeforces.com/problemset/problem/1/A


    
    
  


```python
def theatre_square():
    n,m,a=map(int,input().split()) 
 
    if m%a==0: 
        r1=m//a 
    else: 
        r1=m//a+1 
 
    if n%a==0: 
        r2=n//a 
    else: 
        r2=n//a+1 
 
    print(r1*r2) 
```

```python
import matplotlib.pyplot as plt

def plot_theatre_square(n, m, a):
    x = [0, n, n, 0, 0]
    y = [0, 0, m, m, 0]
    plt.plot(x, y)
    plt.axis('equal')
    plt.show()
    
plot_theatre_square(6, 6, 4)    
```
